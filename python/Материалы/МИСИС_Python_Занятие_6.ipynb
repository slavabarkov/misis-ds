Обратим внимание на то, что все значения в массиве должны быть одного типа:

````python
np.array([
          [1, 2],
          ["a", "b"]
])
````

````python
np.array([1, 2.5])
````

В первом случае все элементы были конвертированы в строки, во втором — в числа с плавающей точкой.

Обратим также внимание на то, что массивы должны иметь прямоугольную форму:

````python
np.array([
          [1, 2],
          [3, 4, 5]
])
````

Предупреждение подсказывает нам, что технически такая возможность есть, но это нужно для довольно специфических случаев:

````python
np.array([
          [1, 2],
          [3, 4, 5]
], dtype=object)
````

В этом случае массив имеет размер 2x1. Каждый элемент является объектом, в даннном примере — списком. Тем же образом можно "снять" ограничения, про которые было сказано выше:

````python
np.array([
          "string",
          np.array([1, 2, 3]),
          [2, 3],
          2.5,
          b"bytestring"
], dtype=object)
````

На самом деле ограничения остаются в силе: массив имеет четко заданную прямоугольную форму 5х1, а все элементы в нем одного типа — object.

## Матричные операции

В машинном обучение многие алгоритмы строятся на матричных операциях. Таких, как: векторное произведение векторов, перемножение матриц, разложение матриц.

dot_product_img.png

Векторное произведение двух векторов численно равно площади образованного ими параллелограмма. В данном случае — прямоугольника.

````python
vector_1 = np.array([0, 2])
vector_2 = np.array([4, 0])

cross_product = np.cross(vector_1, vector_2)
cross_product
````

А скалярное произведение вектора самого на себя — квадрат его длины:

````python
dot_product = np.dot(vector_1, vector_1)
dot_product
````

В нейронных сетях также используются произведение матриц:

````python
matrix_1 = np.array([
    [1, 2, 3],
    [2, 3, 4],
    [5, 6, 7]
])

identity_matrix = np.eye(3)

np.matmul(matrix_1, identity_matrix)
````

Свойство единичной матрицы: EA = AE = A для любой квадратной матрицы А.

В обработке изображений может быть полезным использование матриц поворота:

````python
matrix_1 = np.vstack([vector_1, vector_2])

rotation_matrix = np.array([
    [np.cos(np.pi / 4), np.sin(np.pi / 4)],
    [-np.sin(np.pi / 4), np.cos(np.pi / 4)]
])

np.matmul(matrix_1, rotation_matrix)
````

## Запись и чтение

Очевидный способ записи массива на диск — через текстовое представление.

````python
arr_to_write = np.random.randn(1000, 1000)
````

````python
with open("text_file", "w") as f:
  for row in arr_to_write:
    f.write(",".join([str(f) for f in row]) + "\n")
````

````python
written_arr = np.zeros((1000, 1000))

with open("text_file", "r") as f:
  lines = f.readlines()
  for i, line in enumerate(lines):
    written_arr[i] = [float(s) for s in line.split(",")]
````

````python
(arr_to_write == written_arr).mean()
````

````python
!ls -lhrt
````

Проблема этого способа в том, что 19МБ много для миллиона чисел. Правильнее было бы ожидать около 8МБ (одно число — 64 бита — 8 байта). Более того, в этом случае неясно, как поступать в трех-, четырех- (и более) мерными массивами.

Поэтому разумнее воспользоваться средствами numpy для этого:

````python
np.savez("np_bin_file.npz", arr_to_write)
````

````python
written_arr = np.load("np_bin_file.npz")["arr_0"]
````

````python
(arr_to_write == written_arr).mean()
````

### Скалярное произведение

Попробуем реализовать скалярное произведение с помощью встроенных типов Python и сравним его по скорости с реализацией от numpy

````python
def dot_product(vector_1, vector_2):
  assert len(vector_1) == len(vector_2)

  result = 0

  for el_1, el_2 in zip(vector_1, vector_2):
    result += el_1 * el_2
````

````python
from time import time
````

````python
vector_1 = np.random.randn(10 ** 5)
vector_2 = np.random.randn(10 ** 5)

start_time = time()
for _ in range(10):
  dot_product(vector_1, vector_2)
print(time() - start_time)
````

````python
vector_1 = np.random.randn(10 ** 5)
vector_2 = np.random.randn(10 ** 5)

start_time = time()
for _ in range(10):
  np.dot(vector_1, vector_2)
print(time() - start_time)
````

### Произведение матриц

Проделаем то же самое для операции произведения матриц:

````python
def matmul(matrix_1, matrix_2):
    result = []
    
    for i in range(len(matrix_1)):
        result.append([])
        for j in range(len(matrix_2[0])):
            result[i].append(0)
            
            for k in range(len(matrix_1[0])):
                result[i][j] += matrix_1[i][k] * matrix_2[k][j]
                
    return result
````

````python
N = 100
m_1 = np.random.randn(N, N).tolist()
m_2 = np.random.randn(N, N).tolist()
````

````python
start_time = time()
for _ in range(10):
    matmul(m_1, m_2)

t1 = time() - start_time

start_time = time()
for _ in range(10):
    np.matmul(m_1, m_2)

t2 = time() - start_time

t1 / t2
````
