## Задание 1.
Напишите запрос, который покажет, какую доставку выполнил какой водитель, на каком грузовике он ее выполнил и какому клиенту доставлял. Мы хотим увидеть такой набор полей в следующем порядке:
-	Дата доставки,
-	Название клиента,
-	Город, где находится клиент
-	Имя и фамилию водителя 
-	Марка грузовика

```python
%%sql

SELECT
    sh.ship_date,
    cus.cust_name,
    cus_c.city_name cust_city,
    CONCAT(d.first_name, ' ', d.last_name) driver_name,
    t.make truck_make
FROM
    shipment sh
    LEFT JOIN customer cus ON cus.cust_id = sh.cust_id
    LEFT JOIN city sh_c ON sh_c.city_id = sh.city_id     -- Город, в котором была выполнена доставка
    LEFT JOIN city cus_c ON cus_c.city_id = cus.city_id  -- Город, в котором зарегистрирован клиент
    LEFT JOIN driver d ON d.driver_id = sh.driver_id
    LEFT JOIN truck t ON t.truck_id = sh.truck_id
```

## Задание 2.
Модифицируйте получившийся запрос из Задания 1 (секцию from не менять!) чтобы подсчитать, в какие города осуществлял доставки каждый водитель и сколько доставок он совершил в каждый город. Отсортируйте результаты для каждого водителя в порядке убывания количества рейсов в город.

```python
%%sql

SELECT
    CONCAT(d.first_name, ' ', d.last_name) driver_name,
    sh_c.city_name shipment_city_name,
    COUNT(sh.ship_id) shipments_count
FROM
    shipment sh
    LEFT JOIN customer cus ON cus.cust_id = sh.cust_id
    LEFT JOIN city sh_c ON sh_c.city_id = sh.city_id     -- Город, в котором была выполнена доставка
    LEFT JOIN city cus_c ON cus_c.city_id = cus.city_id  -- Город, в котором зарегистрирован клиент
    LEFT JOIN driver d ON d.driver_id = sh.driver_id
    LEFT JOIN truck t ON t.truck_id = sh.truck_id
GROUP BY
    d.driver_id,
    d.first_name,
    d.last_name,
    sh_c.city_name
ORDER BY
    d.driver_id,
    COUNT(sh.ship_id) DESC
```
