## Задание 1.


Напишите запрос по поставкам, который выдаст нам список водителей, город проживания водителя (название города!), город, куда доставлял груз водитель (тоже название!), номер рейса в данный город по дате и общее количество рейсов данного водителя. 

**Решение:**
Следует уточнить определенные требования у заказчика:

- Требование "список водителей". Какую инофрмацию необходимо выводить о водителе? В данном решении будем выводить имя и фамилию.
- Требование "Номер рейса". Нужно ли выводить номер рейса по всем водителям, или по рассматриваемому водителю? В данном решении будем выводить номер рейса в определенном городе по дате для рассматриваемого водителя.
- Требование "Общее количество рейсов данного водителя". Нужно ли выводить общее количество рейсов в рассматриваемом городе, или независимо от города? В данном решении будем выводить количество независимо от города.

```python
%%sql

SELECT 
    d.first_name,
    d.last_name,
    cd.city_name driver_city_name,
    cs.city_name ship_city_name,
    s.ship_date,
    RANK() OVER ship_id_date route_num,
    COUNT(s.ship_id) OVER driver_stat route_count
FROM 
    shipping.driver d 
    LEFT JOIN shipping.city cd ON cd.city_id = d.city_id 
    RIGHT JOIN shipping.shipment s ON s.driver_id = d.driver_id
    LEFT JOIN shipping.city cs ON cs.city_id = s.city_id
WINDOW 
    ship_id_date AS (PARTITION BY d.driver_id, s.city_id ORDER BY s.ship_date),
    driver_stat AS (PARTITION BY d.driver_id);
```

## Задание 2.


Напишите запрос по книгоиздателям, который покажет нам название книги, автора, жанр, год издания, издательство, общее число изданных книг издателя, средний рейтинг книги данного жанра для издательства.  

```python
%%sql

SELECT 
    b.book_name,
    b.author,
    b.genre,
    b.publishing_year,
    b.publisher,
    COUNT(b.book_id) OVER by_pub pub_books_count,
    (AVG(b.book_average_rating) OVER by_pub_genre)::DECIMAL(5,3) pub_genre_avg_rating
FROM 
    books.books b
WINDOW
    by_pub AS (PARTITION BY b.publisher),
    by_pub_genre AS (PARTITION BY b.publisher, b.genre);
```
